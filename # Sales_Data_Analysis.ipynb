## Exploratory Data Analysis

## Now lets analyze the data based on Gender first

```python
data.columns
```

```python
sns_plot=sns.countplot(x='Gender',data=data)
```

```python
# Now lets show value as well , in the bar to know the exact count of the male and female

sns_plot=sns.countplot(x='Gender',data=data, color='g')
for bars in sns_plot.containers:
    sns_plot.bar_label(bars)
    
```

### Insight: From above plot , it is clearly observed that the female have more contribution in sales number compare to male . Which means that females have made more number of purchase compare to the male

```python
data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
amt=data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Gender',y='Amount',data=amt)
```

### On analysis based on gender , we can conclude females have more purchasing power compare to male 

## Analysis based on Age Group

```python
sns.set(rc={'figure.figsize':(15,5)})
age_plot=sns.countplot(x='Age Group',data=data,hue='Gender')
for bars in age_plot.containers:
    age_plot.bar_label(bars)
```

```python
# total amount 
age_=data.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Age Group',y='Amount',data=age_)
```

### Insight: Age group of 26-35 have more purchasing power . And to be more specific females from Age group 26-35 have made more purchased 

## Analysis based on States

```python
data.columns
```

### Now lets try to know which state has more number of orders so that the state with less number of orders received can be improved by the company 

```python
state_plot=data.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders', data=state_plot)
```

### Insight :From the above plot , we can see that Uttar Pradesh is the state from where the maximum number of orders of the product has been received and also same with the Maharastra and Karnataka as well 

```python
state_plot=data.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x='State',y='Amount', data=state_plot)
```

### Insight: If you observe the above plot carefully and compare it with the previous plot , it is found that though Haryana has less number of orders but it contributes more to the revenue compare to Kerala
